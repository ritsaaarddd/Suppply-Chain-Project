## Insight: High Sales vs Low Stock Levels

The analysis shows that the number of products sold is significantly higher than the available stock levels across all product types.

This indicates strong demand for these products. However, it may also suggest that inventory levels are not sufficient to meet demand, potentially leading to stock shortages.

## Business Implication:
The company may need to increase stock levels or improve inventory replenishment strategies to avoid stockouts.

```python
import pandas as pd
df = pd.read_csv("../data/processed/cleaned_supply_chain_inventory_dataset.csv")
import matplotlib.pyplot as plt

df.groupby('product_type')[['stock_levels', 'number_of_products_sold']].mean().plot(kind='bar')

plt.title("Stock vs Demand by Product Type")
plt.xlabel("Product Type")
plt.ylabel("Average Values")
plt.show()
```
