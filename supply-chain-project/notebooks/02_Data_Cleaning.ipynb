## Data Cleaning Step 1: Column Name Standardization
This step ensures column names are consistent and easy to work with.

```python
import pandas as pd

# Load dataset
df = pd.read_csv("../data/raw/dirty_supply_chain_inventory_dataset.csv")
```

```python
# Standardize column names
df.columns = (
    df.columns.str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

df.columns
```

##  Data Cleaning Step 2: Removing Duplicate Records
This step removes duplicate entries to maintain data integrity.

```python
# Remove duplicates
df = df.drop_duplicates()

# Verify
print("Remaining duplicates:", df.duplicated().sum())
```

## Data Cleaning Step 3: Standardizing Categorical Values
This step fixes inconsistent values in categorical columns.

```python
# Clean transportation modes
df['transportation_modes'] = df['transportation_modes'].str.lower().str.strip()

df['transportation_modes'] = df['transportation_modes'].replace({
    'road': 'Road',
    'ro ad': 'Road',
    'road ': 'Road',
    'air': 'Air',
    'rail': 'Rail',
    'sea': 'Sea'
})

# Check cleaned values
df['transportation_modes'].value_counts()
```

## Data Cleaning Step 4: Handling Missing Values
This step fills missing values using appropriate statistical methods.

```python
# Fill numeric columns with median
for col in df.select_dtypes(include='number').columns:
    df[col] = df[col].fillna(df[col].median())

# Fill categorical columns with mode
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])

# Verify
df.isnull().sum()
```

## Data Cleaning Step 5: Feature Engineering
This step creates new columns for better analysis.

```python
# Inventory turnover ratio
df['inventory_turnover_ratio'] = df['number_of_products_sold'] / (df['stock_levels'] + 1)

# Estimated profit
df['estimated_profit'] = df['revenue_generated'] - df['costs']

df[['inventory_turnover_ratio', 'estimated_profit']].head()
```
